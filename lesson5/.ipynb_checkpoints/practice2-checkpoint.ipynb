```python
score = int(input("請輸入分數"))
if score >= 90:
    grade = '優'
else:
    if score >= 80:
        grade = '甲'
    else:
        if score >= 70:
            grade = '乙'
        else:
            if score >= 60:
                grade = '丙'
            else:
                grade = '丁'

print(grade)
```

```python
#多向選擇
score = int(input("請輸入分數"))
if score >= 90:
    grade = '優'
elif score >= 80:
    grade = '甲'
elif score >= 70:
    grade = '乙'
elif score >= 60:
    grade = '丙'
else:
    grade = '丁'

print(grade)
```

## 異常處理 (try-except) 與資料驗證

```python
money = int(input("請輸入金額:"))
print("輸入的金額為:", money)
```

```python
try:
    money = int(input("請輸入金額:"))
    print("輸入的金額為:", money)
except ValueError:
    print("輸入格式錯誤，請輸入整數數字！")
except:
    print("不知明的錯誤")
```

```python
# 國文100分同時數學100分,獎金1000元
# 2科只有1個100分,獎金500
# 如果都沒100,獎金0元

bonus = 0
chinese = 100
math = 100
if chinese == 100:
    if math == 100:
        bonus = 1000
    else:
        bonus = 500
elif math == 100:
    bonus = 500
    
```

```python
# 國文100分同時數學100分,獎金1000元
# 2科只有1個100分,獎金500
# 如果都沒100,獎金0元


chinese = 100
math = 100
if chinese == 100 and math == 100:
    bonus = 1000
elif chinese == 100 or math == 100:
    bonus = 500
else:
    bonus = 0

print(bonus)
```
