```python
#雙向選擇
a = 13
if a >= 14:
    a -= 2
    print(a)
else:
    a += 2
    print(a)
```

```python
#單向選擇
age = 15
if age >= 18:
    print("我已經可以投票")
print("應用程式結束")
```

```python
#巢狀判斷
furry = False
small = True

if furry:
    print("這是有毛的")
    if small:
        print("這是小的")
        print("它是貓")
    else:
        print("這是大的")
        print("它是熊")
else:
    print("這是代表沒有毛")
    if small:
        print("這是小的")
        print("它是小蜥蜴")
    else:
        print("這是大的")
        print("它是人類或是沒毛的熊")
```
