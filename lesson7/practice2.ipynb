```python
#list的typehint
#list
rabbits:list[str] = ["Flopsy", "Mopsy", "Cottontail", "Peter"]


for rabbit in rabbits:
    print(rabbit)
```

```python
#dict的typehint

country_code:dict[str,str] = {"tw":"Taiwan", "jp": "Japan", "hk":"Hong Kong", "kr": "Korea"}
for code in country_code:
    print(f"key:{code}")
    print(f"value:{country_code[code]}")
    print("=============")
```

```python
#輸入英文字母 (按 q 離開)
while True:
    stuff = input("請輸入小寫英文字[按q會離開]:")
    if stuff == 'q':
        break
    print(stuff.capitalize())
print("程式結束")
```

```python
#累加正偶數
#設計一個程式讓使用者輸入數值，只有正偶數值會被加總，若輸入奇數則跳過，輸入負數則結束程式：
total:int = 0
num:int = 0
while True:
    input_value = int(input(f"請輸入第{num + 1}個數值:"))    
    if input_value < 0:
        break
        
    num += 1

    if input_value % 2 == 1:
        continue

    total += input_value

print(f"輸入的次數是{num}, ，所有輸入的正偶數的加總是:{total}")
```
